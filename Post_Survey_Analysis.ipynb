```python
def plotbar(m,f,cm,cf):
 x = np.arange(5)
 width = 0.35

 fig, ax = plt.subplots()
 mr = ax.bar(x - width/2, m, width, label='Male',color=cm)
 fr = ax.bar(x + width/2, f, width, label='Female',color=cf)

 # Add some text for labels, title and custom x-axis tick labels, etc.
 ax.set_ylabel('Number of responses')
 ax.set_title('Number of responses by gender')
 plt.xticks(x, ["Never","Almost Never","Sometimes","Fairly Often","Very Often"],rotation=45)
 ax.legend()

 fig.tight_layout()

 plt.show()
```

```python
def seggregateResponses(qs,qs_df,i):
 options = sorted(list(dict(qs).keys()))
 temp = qs_df["Question "+str(i)].value_counts()
 k = dict(temp).keys()
 options
 m = []
 f = []
 for i in options:
  if ("Male",i)  in k:
     m.append(temp["Male",i])
  else:
     m.append(0)
     
  if ("Female",i)  in k:
     f.append(temp["Female",i])
  else:
     f.append(0)
 return m,f
```

## Evaluation of the Respondent's Stress Levels
  In this section, we will evaluate the respondent stress level based on the following criteria :

*   Each question in this section had the following options:
         1. Never - 0
         2. Almost Never - 1
         3. Sometimes - 2
         4. Often - 3
         5. Very Often - 4
*   At first, we will categorise a sentence on the basis of its nature which can be either negative or positive with respect to its meaning.
*   If a sentence is negative then 0 is considered to be good and will be assigned 4 during evaluation.
*   If the sentence is positive then 0 is considered bad and 4 will be considered good.
*   The score will be the sum of all the responses given by the respondent.




```python
negative = [4,5,7,8]
def evaluateRespondent(row,n):
    score = 0
    for i in range(1,11):
      if i in negative:
        score += n - int(row["Question "+str(i)][0])
      else:
        score += int(row["Question "+str(i)][0])
    return score
```

```python
s = df1.apply(evaluateRespondent,args=[4],axis=1)
sns.histplot(s,color='crimson',edgecolor='black',kde=True)
plt.title("Histogram of Stress Levels")
plt.xlabel("Stress Level")
plt.ylabel("Probability")
plt.show()
```

```python
norm.fit(s)
```

```python
def classifyIntoCategory(stressLevel):
  if stressLevel >= 0 and stressLevel <= 13:
    return 'Low'
  elif stressLevel >=14 and stressLevel <=26:
    return 'Moderate'
  else:
    return 'High'
```

```python
cs = s.apply(classifyIntoCategory)
vcs = cs.value_counts()
plt.barh(sorted(list(dict(vcs).keys())),dict(vcs).values(),color=["aqua","crimson","yellow"],edgecolor="black")
plt.yticks([2,1,0],["Low","Moderate","High"],rotation=45)
plt.xlim([0,65])
plt.title("Stress Category v/s No.of respondents in that category")
plt.xlabel("Number of respondents")
plt.show()
```

## Evaluation of the Respondent's Stress Levels

```python
negative = [1,2,3,5]
```

```python
def classifyIntoCategory(stressLevel):
  if stressLevel >= 0 and stressLevel <= 8:
    return 'Low'
  elif stressLevel >=9 and stressLevel <=17:
    return 'Moderate'
  else:
    return 'High'
```

# Computation of the Overall Stress Levels of the Respondents

```python
sns.kdeplot(r+s)
plt.title("Overall Stress Levels Analysis")
plt.xlabel("Stress Level(Overall)")
plt.ylabel("Probability")
plt.show()
```

```python
norm.fit(r+s)
```

```python
def classifyIntoCategory(stressLevel):
  if stressLevel >= 0 and stressLevel <= 21:
    return 'Low'
  elif stressLevel >=22 and stressLevel <=42:
    return 'Moderate'
  else:
    return 'High'
```

```python
cs = (r+s).apply(classifyIntoCategory)
vcs = dict(cs.value_counts())
if "Low" not in vcs.keys():
  vcs["Low"] = 0
if "Moderate" not in vcs.keys():
  vcs["Moderate"] = 0
if "High" not in vcs.keys():
  vcs["High"] = 0
plt.barh(sorted(list(vcs.keys())),vcs.values(),color=["crimson","orange","yellow"],edgecolor="black")
plt.yticks([2,1,0],["Low","Moderate","High"],rotation=45)
plt.xlim([0,65])
plt.title("Overall Stress Category v/s No.of respondents in that category")
plt.xlabel("Number of respondents")
plt.show()
```
